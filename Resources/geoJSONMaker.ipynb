```python
import pandas as pd
import geopandas as gpd
import json
import geojson
```

```python
counties_df = gpd.read_file("counties.json", encoding="utf-8")
counties_df
```

```python
info_df = pd.read_csv("BreweryProject2-RGbranch/countydemographicsv1.csv", encoding = "utf-8")
info_df
```

```python
ordered_counties = counties_df.sort_values(by=["GEO_ID"]).reset_index(drop = True)
ordered_counties
```

```python
useful_info_df = info_df[["median_HHI_2019", "bach_deg_pct_14to18"]]
useful_info_df
```

```python
final_county_df = pd.concat([useful_info_df, ordered_counties], axis = 1)
final_county_df
```

```python
final_county_gdf = gpd.GeoDataFrame(final_county_df)
```

```python
final_county_gdf.to_file("final_counties.geojson", driver ='GeoJSON', encoding = "utf-8")
```
