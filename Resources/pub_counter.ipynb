```python
import pandas as pd
```

```python
brew_path = "breweries_v1.csv"
county_path = "countydemographicsv1latlong.csv"
brew_df = pd.read_csv(brew_path)
county_df = pd.read_csv(county_path)
brew_df
```

```python
grouped_df=brew_df.groupby(["CountyID"]).size().reset_index(name='Brewery_Count')
grouped_df
```

```python
grouped_df = grouped_df.rename(columns={"CountyID":"fips_txt"})
grouped_df
```

```python
merged_df = pd.merge(county_df, grouped_df, how="outer")
merged_df
```

```python
merged_df["Brewery_Count"] = merged_df["Brewery_Count"].fillna(0)
merged_df
```

```python
merged_df.to_csv("FinalCountyDemographics.csv")
```
