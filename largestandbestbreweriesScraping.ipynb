```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from re import search

from urllib.request import Request, urlopen

```

```python
url_largest = 'https://www.brewersassociation.org/press-releases/brewers-association-announces-top-50-brewing-companies-by-sales-volume-of-2019/'
url_best = 'https://www.americancraftbeer.com/ratebeers-top-100-breweries-2019/'

req = Request(url_best, headers={'User-Agent':'Chrome/87.0.4280.141'})
best_rawhtml = urlopen(req).read()
```

```python
# pull table version of largest craft breweries
largest_table = pd.read_html(url_largest)
```

```python
largestbreweries_df = largest_table[0]
largestbreweries_df = largestbreweries_df.rename(columns={'Brewery Name': 'BreweryName'})


largestlist = []
for index in largestbreweries_df['BreweryName']:
    largestlist.append(True)

largestbreweries_df["Top50Largest"] = largestlist

largestbreweries_df
```

```python
best_table = pd.read_html(best_rawhtml)
best_table
```

```python
bestbreweries_df = best_table[0]
bestbreweries_df = bestbreweries_df.rename(columns={'Name': 'BreweryName'})

usbestlist = []
bestlist = []
for index in bestbreweries_df['BreweryName']:
    bestlist.append(True)
    if "United States" in index:
        usbestlist.append(True)
    else:
        usbestlist.append(False)

bestbreweries_df["Top50Best"] = bestlist
bestbreweries_df["inUSA"] = usbestlist

bestbreweries_df
```

```python
bestUSbreweries_df = bestbreweries_df.loc[bestbreweries_df["inUSA"] == True, :].head(50).reset_index()

bestUSbreweries_df = bestUSbreweries_df[["BreweryName","Top50Best"]]

bestbrewnames = []
bestbrewstates = []

for index in bestUSbreweries_df['BreweryName']:
    flagsplit = index.split('ðŸ‡ºðŸ‡¸')
    bestbrewnames.append(flagsplit[0])
    commasplit = index.split(',')
    bestbrewstates.append(commasplit[1])
    
bestUSbreweries_df["BreweryName2"] = bestbrewnames
bestUSbreweries_df["State"] = bestbrewstates

bestUSbreweries_df = bestUSbreweries_df[["BreweryName2",'State',"Top50Best"]]
bestUSbreweries_df['BreweryName'] = bestUSbreweries_df['BreweryName2']

bestUSbreweries_df = bestUSbreweries_df[["BreweryName",'State',"Top50Best"]]
bestUSbreweries_df
```
